```python
from astropy.io import fits
from matplotlib import pyplot as plt
import numpy as np
import astropy.units as u
```

```python
# density probe check
with fits.open('snapshot_150_densprobe_dust_rho_xy.fits') as hdul:
    hdul.info()
    data = hdul[0].data
    header = hdul[0].header
print(repr(header))
# going from msun/au2 to msun/pc2
au_to_cm = u.AU.to('pc')
conv_factor = 1 / (au_to_cm)**2
dust_density_xy = data * conv_factor
im = plt.imshow(np.log10(dust_density_xy), origin='lower', cmap='viridis')
plt.xlim(500,530)
plt.ylim(500,530)
print(f'Min value is: {np.min(dust_density_xy)}')
print(f'Max value is: {np.max(dust_density_xy)}')
print(f'Min logged value is: {np.min(np.log10(dust_density_xy))}')
print(f'Max logged value is: {np.max(np.log10(dust_density_xy))}')
plt.xlabel('X (pc)')
plt.ylabel('Y (pc)')
plt.title(f'Dust Density Probe \n(Snapshot 150)')
cbar = plt.colorbar(im, label=f'$\Sigma_{{gas}} (M_{{\odot}}pc^{-2})$')
```

```python
# Open the FITS file
hdul = fits.open('snapshot_150_O1_total.fits')
data = hdul[0].data
print('Data shape:', data.shape)

for i in range(data.shape[0]):
    slice_i = data[i, :, :]
    if np.all(slice_i == 0):
        print(f'Slice {i} is all zeros.')
    else:
        print(f'Slice {i} has nonzero values.')
    plt.figure()
    plt.imshow(slice_i, origin='lower')
    plt.title(f'Slice {i}')
    plt.show()
```
